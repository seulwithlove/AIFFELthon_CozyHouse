```python
import torch
import cv2
from ultralytics import YOLO
import torch
import torch.nn as nn
from torchvision.models import resnet50
from torchvision.models.detection import keypointrcnn_resnet50_fpn
from torchvision.models.detection import KeypointRCNN
```

```python
class YOLOv8Combined(nn.Module):
    def __init__(self, segmentation_model_path, pose_model_path):
        super(YOLOv8Combined, self).__init__()
        # YOLOv8 Segmentation 모델 불러오기
        self.segmentation_model = YOLO(segmentation_model_path)
        self.segmentation_model.to('cuda')

        # YOLOv8 포즈 추정 모델 불러오기
        self.pose_model = YOLO(pose_model_path)
        self.pose_model.to('cuda')

    def forward(self, image):
        # 세그멘테이션 수행
        segmentation_results = self.segmentation_model(image)

        # 포즈 추정 수행
        pose_results = self.pose_model(image)

        return segmentation_results, pose_results

# 병합된 모델 인스턴스화
segmentation_model_path = 'yolov8x-seg.pt'
pose_model_path = 'yolov8x-pose-p6.pt'
combined_model = YOLOv8Combined(segmentation_model_path, pose_model_path)

# 입력 데이터 로드
input_data = "data/lee_data.mp4"  # 입력 비디오 경로

# 추론 수행
segmentation_results, pose_results = combined_model(input_data)
```
